# BSHMM : Model test drive

## 1. Model initialization

The following script initializes the HMM class and sets the basic parameters.

The methylation level states are bound with values close to 0 and 1, to avoid divide-by-zero errors.
The initial transition matrix is inspired from empirical data visualization. We contraint it to have values i=j higher than the rest of the matrix, meaning that the methylation profile tends to stay more often at the same level rather than often changing from high to low values. We also make sure the matrix is row-stochastic, the HMM class has a function to check that the paramters truly describe probabilities. 

Finally, the initial distribution is set to a uniform distribution. This deosn't have a high influence since we usually work on a single long sequence. 

We then create an instance of the model with the 3 parameters described above.

```python
import hmm
%load_ext autoreload
%autoreload 2

states = np.array([.0001, .1, .2, .3, .4, .5, .6, .7, .8, .9, .9999])
N = len(states)

initial = np.array([1] * N, dtype=np.float32)
initial /= sum(initial)


transition = np.eye(N) + .01
for i in xrange(len(transition)):
	transition[i] /= sum(transition[i])

model = hmm.Hmm(states, initial, transition)
```

## 2. Model-driven data simulation (Generative HMM)

We have implemented a generative model of our HMM to simulate data based on a set of given parameters. The simulation happens in three steps, successively generating a dataset for the genome coverage, the cytosine counts and the methylation profile. 

While C counts and methylation profiles are entirely based on the Markov chain and the paramters set, the coverage has to be simulated apart. We have implemented three different simulation models :  
* Fixed homogenous coverage, set at a given value, the simplest model  
* Poisson-distributed coverage.  
* Sampled at each position using a floating window. We use the immediate neighborhooud distribution (of a given size) to resample the coverage value at every position. This is intented to mimic coverage variability as much as possible.

The following snippet imports real data from a samtools pileup file, and also runs the simulation routine. 
I have intentionnally included all the arguments, though for the "fixed" coverage only fix_cov_val is required. The arg signification is pretty much self explanatory.
PS : The routine always takes a real coverage dataset too. This might be changed to become mandatory for the "window" and "real" coverage type only. 

```python
# length of the simulated sequence
L = 800

# Real data imported from a samtools pileup file
# pos : Cytosine position in the genome
# cov : Coverage depth
# count : C counts (ie. number of aligned cytosines at pos)
meth_data = np.genfromtxt('meth.dat', 
	usecols = (0, 2, 4),
	dtype=[('pos','i8'),('cov','i8'), ('count','f8')])[0:L]

# Simulation routine
path, obs, cov = model.generateData(meth_data['cov'],
	cov_type="fixed",
	seq_length = 2000,
	window_size = 40,
	lam = 15,
	fix_cov_val=100
)
```

## 3. Decoding algorithms validation

The coverage variability is the main source of bias and lack of precision in finding the best methylation profile. To make sure our model is resilient to coverage variability, we will run the decoding algorithm and compare the accuracy of the results. We will set the coverage to different fixed values (ranging from 5 to 100), and then go on with Poisson, floating window sampled and real coverage.
(Confidence intervals are to be re-implemented, it behaves pretty strangely for now.)

###a. Posterior decoding

#### Overview

```python
path, obs, cov = model.generateData(meth_data['cov'], cov_type="fixed", seq_length = L, fix_cov_val=30)
best_path, acc = model.decode(meth_data['pos'], obs, cov, method="posterior", graph=True, real_path = path, ci_alpha = .80)

path, obs, cov = model.generateData( meth_data['cov'], cov_type="poisson", seq_length = L, lam=10)
best_path, acc = model.decode(meth_data['pos'], obs, cov, method="posterior", graph=True, real_path = path, ci_alpha = .5)

path, obs, cov = model.generateData(meth_data['cov'], cov_type="window", seq_length = L, window_size=40)
best_path, acc = model.decode(meth_data['pos'], obs, cov, method="posterior", graph=True, real_path = path, ci_alpha = .5)

path, obs, cov = model.generateData(meth_data['cov'], cov_type="real", seq_length = L)
best_path, acc = model.decode(meth_data['pos'], obs, cov, method="posterior", graph=True, real_path = path, ci_alpha=.5)
```

#### Posterior decoding accuracy vs. Coverage parameters 
The following snippet shows how the accuracy varies with different parameters values, for each of the three coverage types.
(The execution might take some time, please be patient).


```python
fixed_accs = []
for val in xrange(2, 60):
	path, obs, cov = model.generateData(meth_data['cov'], cov_type="fixed", seq_length = 800,fix_cov_val = val)
	best_path, acc= model.decode(meth_data['pos'], obs, cov,method="posterior",graph=False,real_path = path, ci_alpha=.50)
	fixed_accs.append(acc)

fig = figure(figsize=(16,5), dpi=100)
subplot(131)
plot(range(2,60), fixed_accs, '+')
title('Acc. vs. various fixed coverage depths')
xlabel("Coverage depth")
ylabel("Accuracy")


poisson_accs = []
for val in xrange(2, 30):
    path, obs, cov = model.generateData(meth_data['cov'], cov_type="poisson", seq_length = 800,lam = val)
    best_path, acc = model.decode(meth_data['pos'], obs, cov,method="posterior",graph=False,real_path = path, ci_alpha=.50)
    poisson_accs.append(acc)

subplot(132)
plot(range(2,30), poisson_accs, '+')
title("Acc. vs. Poissons' lambda")
xlabel("Lambda")
ylabel
window_accs = []
for val in xrange(10, 200, 5):
    
	path, obs, cov = model.generateData(meth_data['cov'], cov_type="window", seq_length = 800, window_size=val)
	best_path, acc= model.decode(meth_data['pos'], obs, cov,method="posterior",graph=False,real_path = path, ci_alpha=.50)
	window_accs.append(acc)

subplot(133)
plot(range(10,200,5), window_accs, '+')
title("Acc. vs. neighborhood resampling")
xlabel("Floating window size")

show()
```
